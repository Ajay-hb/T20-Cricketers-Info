# **BattingSummary**

```python
with open('/content/sample_data/t20_wc_batting_summary.json') as f:
  data=json.load(f)
  all_records=[]

  for rec in data:
    all_records.extend(rec['battingSummary'])
all_records
df_batting=pd.DataFrame(all_records)
df_batting.head(11)
```

```python
df_batting['out/not out']=df_batting.dismissal.apply(lambda x:"out" if len(x)>0 else "not out")
```

```python
df_batting.head()
```

```python
df_batting.drop(columns=['dismissal'],inplace=True)
```

```python
df_batting.head(11)
```

```python
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x:x.replace('â€',''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x:x.replace('†', ''))
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x:x.replace('\xa0',''))
```

```python
df_batting.head(11)
```

```python
match_ids_dict={}

for index,row in df_match.iterrows():
  key1=row['team1']+' vs '+row['team2']
  key2=row['team2']+' vs '+row['team1']

  match_ids_dict[key1]=row['match_id']
  match_ids_dict[key2]=row['match_id']

match_ids_dict
```

```python
df_batting['match_id']=df_batting['match'].map(match_ids_dict)
df_batting.head()
```

```python
 df_batting.to_csv('/content/t20_csv_files/fact_batting_summary.csv',index=False)
```

# **Bowling Summary**

```python
with open('/content/sample_data/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
all_records[:2]
```

```python
df_bowling = pd.DataFrame(all_records)
print(df_bowling.shape)
df_bowling.head()
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_ids_dict)
df_bowling.head()
```

```python
import os

if not os.path.exists('t20_csv_files'):
    os.makedirs('t20_csv_files')
```

```python
df_bowling.to_csv('t20_csv_files/fact_bowling_summary.csv', index = False)
```
